````python
import pandas as pd

company_data_active = (
    data_loader.load_company_data(closed=False)
    .assign(Closed=False)
)
company_data_closed = (
    data_loader.load_company_data(closed=True)
    .assign(Closed=True)
)
company_data = pd.concat([company_data_active, company_data_closed], axis=0)

column_intersection = list(set(meta_data['영문칼럼명'].tolist()).intersection(set(company_data.columns.tolist())))
column_mapping = meta_data.loc[lambda x : x['영문칼럼명'].isin(column_intersection), ['영문칼럼명', '한글칼럼명']].set_index('영문칼럼명').to_dict()['한글칼럼명']

company_data = (
    company_data.rename(columns=column_mapping)
)
````
